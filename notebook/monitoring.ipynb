# Loading data from MYSQL database

```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql

# MySQL connection details
username = "root"
password = "akukanem_12"
host = "localhost"        # or IP address
port = 3306               # default MySQL port
database = "mlops"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Read 'predictions' table
predictions_df = pd.read_sql("SELECT * FROM predictions", engine)

# Read 'machine_data_1' table
machine_data_df = pd.read_sql("SELECT * FROM machine_data_1", engine)
```

```python
reference_df = machine_data_df.drop(['id','product_id','twf','hdf','pwf','osf','rnf'], axis=1)
```

```python
predictions_df.head()
```

```python
current_df = predictions_df.drop(["id","prediction_label","timestamp"],axis=1)
```

```python
current_df.loc[:, 'type_val'] = current_df['type_val'].replace({0:"L", 1:"M", 2:"H"})
```

```python
reference_df["Power"] = reference_df["torque_nm"] * reference_df["rotational_speed_rpm"]
```

```python
col_to_move = 'Power'

# Reorder so the chosen column is last
current_df = current_df[[c for c in current_df.columns if c != col_to_move] + [col_to_move]]
```

```python
current_df.head()
```

# Model Drift with EvidentlyAI(This when ground truth is not available)

```python
from evidently.report import Report
from evidently.metrics import DataDriftPreset

# Run Evidently
report = Report(metrics=[DataDriftPreset()])
report.run(reference_data=reference_df, current_data=current_df)

# Extract results as dictionary
result = report.as_dict()

# Get drift percentage
share_drifted = result["metrics"][0]["result"]["dataset_drift"]["share_of_drifted_columns"]
print(f"Share of drifted columns: {share_drifted:.2f}")
```
